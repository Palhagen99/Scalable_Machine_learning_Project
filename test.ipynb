```python
#Test of predicting stock price
import ModelForecaster as mf

prediction, prev_closing = mf.forecast_stock_price("AAPL")

```

```python
#Plot the close price and then at the end of then plot the prediction in a different color
import matplotlib.pyplot as plt
import numpy as np

plt.plot(prev_closing, label='Previous Closing Prices', linestyle='--',)

# Create an array of indices for the predicted values, right after the last index of prev_closing
predicted_indices = np.arange(len(prev_closing), len(prev_closing) + len(prediction))

# Plot the predicted closing prices in red, using the correct indices
plt.plot(predicted_indices, prediction, color='red', label='Predicted Prices',linestyle='--',)

# Adding labels and a legend
plt.xlabel('Days')
plt.ylabel('Price')
plt.title('Stock Prices with Predictions')
plt.legend()

print(prediction)
```

```python
def is_valid_ticker(ticker):
    try:
        # Fetch data for the ticker
        stock = yf.Ticker(ticker)
        info = stock.info

        # Check if the stock has valid information
        if "symbol" in info and info["symbol"] == ticker.upper():
            return True
        return False
    except Exception:
        return False
    
print(is_valid_ticker('AAPL'))
```
